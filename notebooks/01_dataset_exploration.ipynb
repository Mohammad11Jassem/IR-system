```python
import ir_datasets

dataset_name = "antique"
dataset = ir_datasets.load(dataset_name)

print("Dataset:", dataset_name)
print("-" * 50)

print("هل يحتوي على مستندات؟", dataset.has_docs())
print("هل يحتوي على استعلامات؟", dataset.has_queries())
print("هل يحتوي على أحكام صلة؟", dataset.has_qrels())
print("هل يحتوي على نتائج مرتبة مسبقًا؟", dataset.has_scoreddocs())
print("هل يحتوي على أزواج تدريب؟", dataset.has_docpairs())
print("هل يحتوي على سجلات استعلامات؟", dataset.has_qlogs())

if dataset.has_docs():
    print("عدد المستندات:", dataset.docs_count())

if dataset.has_queries():
    print("عدد الاستعلامات:", dataset.queries_count())

if dataset.has_qrels():
    print("عدد أحكام الصلة:", dataset.qrels_count())
```
